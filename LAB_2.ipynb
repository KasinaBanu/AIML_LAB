```python
year=range(2000,2008)
apples=[0.895,0.98,0.456,0.765,0.97,0.456,0.982,0.874]
oranges=[0.987,0.956,0.972,0.567,0.875,0.981,0.764,0.94]
plt.plot(year,apples)
plt.plot(year,oranges)
plt.xlabel('year')
plt.ylabel('yield')
plt.title('crop yields in kanto')
plt.legend(['Apples','Oranges']);
```

```python
plt.plot(year,apples,'s-b')
plt.plot(year,oranges,'o--r')
plt.xlabel('year')
plt.ylabel('yield')
plt.title('crop yields in kanto')
plt.legend(['Apples','Oranges']);
```
